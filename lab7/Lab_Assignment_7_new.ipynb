```python
# 1.
import pandas as pd

# Load the CSV file
file_path = 'yachts_2.csv'
yachts_data = pd.read_csv(file_path)

# Display general information about the data
print("Data Info:")
yachts_data.info()

# Display the first few rows of the data
print("\nFirst 5 Rows:")
print(yachts_data.head())
```

```python
# 2
import pandas as pd

# Load the booking CSV files
bookings_file_path = 'bookings.csv'
bookings_data = pd.read_csv(bookings_file_path, encoding='latin1')

# Filter bookings to keep only those with correctly formatted Yacht_IDs (two characters followed by three numbers)
correct_format = bookings_data['Yacht_ID'].str.match(r'^[A-Z]{2}\d{3}$', na=False)
filtered_bookings_data = bookings_data[correct_format]

# Merge the filtered bookings data with the yachts data on 'Yacht_ID'
merged_data = pd.merge(filtered_bookings_data, yachts_data, left_on='Yacht_ID', right_on='Yacht_Ref', how='inner')

# Display the first few rows of the merged data to check the result
print("Merged Data:")
print(merged_data.head())
```

```python
# 3.
# a.
# Define the desired column order
desired_columns = [
    'Title', 'First_Name', 'Last_Name', 'Street_Address', 'Area', 'Post_Code', 
    'Telephone', 'Yacht_ID', 'Yacht_Name', 'Yacht_Make', 'Yacht_Ref', 
    'Depart_Port', 'Depart_Country'
]

# Reorder the columns of the merged data
ordered_merged_data = merged_data[desired_columns]

# Display the first few rows of the ordered merged data to check the result
print("Ordered Merged Data:")
print(ordered_merged_data.head())

```

```python
# b.
# Define the desired column order including numerical columns
desired_columns_with_numbers = [
    'Title', 'First_Name', 'Last_Name', 'Street_Address', 'Area', 'Post_Code', 
    'Telephone', 'Cust_ID', 'Yacht_ID', 'Yacht_Name', 'Yacht_Make', 'Yacht_Ref', 
    'Depart_Port', 'Depart_Country', 'Built', 'Length(m)', 'Cabins', 'Toilets', 
    'Guests', 'Day_Rate'
]

# Reorder the columns of the merged data
ordered_merged_data = merged_data[desired_columns_with_numbers]

# Ensure the specified columns are of numerical type
ordered_merged_data[['Cust_ID', 'Built', 'Length(m)', 'Cabins', 'Toilets', 'Guests', 'Day_Rate']] = ordered_merged_data[['Cust_ID', 'Built', 'Length(m)', 'Cabins', 'Toilets', 'Guests', 'Day_Rate']].apply(pd.to_numeric)

# Display the first few rows of the ordered merged data to check the result
print("Ordered Merged Data with Correct Data Types:")
print(ordered_merged_data.head())
```

```python
# c.

# Define the desired column order including numerical columns and Depart_Date
desired_columns_with_numbers_and_date = [
    'Title', 'First_Name', 'Last_Name', 'Street_Address', 'Area', 'Post_Code', 
    'Telephone', 'Cust_ID', 'Yacht_ID', 'Yacht_Name', 'Yacht_Make', 'Yacht_Ref', 
    'Depart_Port', 'Depart_Country', 'Depart_Date', 'Built', 'Length(m)', 'Cabins', 
    'Toilets', 'Guests', 'Day_Rate'
]

# Reorder the columns of the merged data
ordered_merged_data = merged_data[desired_columns_with_numbers_and_date]

# Ensure the specified columns are of numerical type
ordered_merged_data[['Cust_ID', 'Built', 'Length(m)', 'Cabins', 'Toilets', 'Guests', 'Day_Rate']] = ordered_merged_data[['Cust_ID', 'Built', 'Length(m)', 'Cabins', 'Toilets', 'Guests', 'Day_Rate']].apply(pd.to_numeric)

# Ensure the Depart_Date column is of date type with the format dd/mm/yyyy
ordered_merged_data['Depart_Date'] = pd.to_datetime(ordered_merged_data['Depart_Date'], format='%d/%m/%Y')

# Display the first few rows of the ordered merged data to check the result
print("Ordered Merged Data with Correct Data Types:")
print(ordered_merged_data.head())
```

```python
import pandas as pd

# Load the CSV file
file_path = 'income_data.csv'
income_data = pd.read_csv(file_path)

# Display basic information about the dataset
income_data_info = income_data.info()
income_data_head = income_data.head()

income_data_info, income_data_head
```

```python
import scipy.stats as stats

# Create a contingency table for race and education
contingency_table = pd.crosstab(income_data['race'], income_data['education'])

# Perform the Chi-square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

chi2, p
```

Statistical Significance: Since the p-value is extremely small, it is far below any common threshold for significance. This means we reject the null hypothesis.

The rejection of the null hypothesis suggests that there is a statistically significant relationship between the variables 'race' and 'education' in the dataset. In other words, the distribution of education levels is not independent of race; some education levels are more or less common among different racial groups than would be expected if there were no association.

```python
# Overtime can be considered an individual that works more than the standard full-time hours (40 hours per week)
# Create an indicator feature for working more than 40 hours per week
income_data['overtime'] = income_data['hours-per-week'] > 40

# Display the first few rows to verify the new feature
income_data.head()
```

```python
# The level of education might influence the number of hours an individual works.
# Create an interaction feature for educational-num and hours-per-week
income_data['education_hours_interaction'] = income_data['educational-num'] * income_data['hours-per-week']

# Display the first few rows to verify the new feature
income_data.head()
```
